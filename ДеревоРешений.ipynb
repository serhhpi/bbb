```python
#дерево решений
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Данные
X = [[25, 1, 60], [18, 0, 100], [20, 1, 80]]  # Температура и погода (1=Солнечно, 0=Дождь, 3=Влажность)
y = [1, 0, 1]  # 1=Гулять, 0=Нет
```

```python
# Обучение модели
model = DecisionTreeClassifier(criterion="gini", max_depth=3)
model.fit(X, y)
```

```python
# Визуализация дерева
plt.figure(figsize=(10, 8))
tree.plot_tree(model, feature_names=["Температура", "Погода","Влажность"], class_names=["Не гулять", "Гулять"],
                filled=True,
    rounded=True
               )
plt.show()
```

```python
#Пример предсказания для новых данных:
new_data = [[18, 1, 100]]  # Температура=22, Погода=Солнечно, Влажность=65
print(model.predict(new_data))  # [1] (Гулять)
```
