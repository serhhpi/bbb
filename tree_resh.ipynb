```python
#дерево решений
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Данные
X = [[25, 1], [18, 0], [20, 1]]  # Температура и погода (1=Солнечно, 0=Дождь)
y = [1, 0, 1]  # 1=Гулять, 0=Нет
```

```python
# Обучение модели
model = DecisionTreeClassifier(criterion="gini", max_depth=2)
model.fit(X, y)
```

```python
# Визуализация дерева
plt.figure(figsize=(10, 6))
tree.plot_tree(model, feature_names=["Температура", "Погода"], class_names=["Нет", "Да"])
plt.show()
```
